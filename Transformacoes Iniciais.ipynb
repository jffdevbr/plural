#### Gênero, Gênero Social e Identidade de Gênero

```python
dados_funcionarios['Gênero'] = dados_funcionarios['Gender']
dados_funcionarios.loc[dados_funcionarios['Gênero']=='Transgender'].groupby('Pronouns').size()
```

```python
dicionario_genero = {
    'Transgender': 'Trans',
    'Male': 'Homem Cisgênero',
    'Female': 'Mulher Cisgênero',
    'Prefer not to say': 'Prefere Não Informar',
    'Non-binary/non-conforming' : 'Não Binário',
    'Other': 'Outro'
}
dados_funcionarios['Gênero'] = dados_funcionarios['Gender'].map(dicionario_genero)

# Visto que todas as pessoas trans utilizam pronomes neutros, será adotado um critério aleatório para definir se será homem ou mulher trans
pessoas_trans = dados_funcionarios['Gender'] == 'Transgender'
dados_funcionarios.loc[pessoas_trans, 'Gênero'] = np.random.choice(
    ['Mulher Transgênero', 'Homem Transgênero'],
    size=pessoas_trans.sum()
)
dados_funcionarios.loc[pessoas_trans][['Name', 'Gender', 'Gênero']].sample(5, random_state=100)
```

```python
# Criar coluna de Gênero Social
mapa_genero_social = {
    'Homem Cisgênero': 'Homem',
    'Homem Transgênero': 'Homem',
    'Mulher Cisgênero': 'Mulher',
    'Mulher Transgênero': 'Mulher',
    'Não Binário': 'Não Binário',
    'Outro': 'Outro',
    'Prefere Não Informar': 'Prefere Não Informar'
}

dados_funcionarios['Gênero Social'] = dados_funcionarios['Gênero'].map(mapa_genero_social)


# Criar coluna de Identidade de Gênero
mapa_identidade = {
    'Homem Cisgênero': 'Cisgênero',
    'Mulher Cisgênero': 'Cisgênero',
    'Não Binário': 'Outro',
    'Outro': 'Outro',
    'Homem Transgênero': 'Transgênero',
    'Mulher Transgênero': 'Transgênero',
    'Prefere Não Informar': 'Prefere Não Informar'
}

dados_funcionarios['Identidade Gênero'] = dados_funcionarios['Gênero'].map(mapa_identidade)
```

```python
# Adicionando Grau de Minoria
dados_funcionarios.loc[(dados_funcionarios['Gênero Social']=='Mulher') | (dados_funcionarios['Gênero Social']=='Não Binário'), 'Grau Minoria'] += 1
dados_funcionarios.loc[(dados_funcionarios['Identidade Gênero']=='Transgênero'), 'Grau Minoria'] += 1
```

#### Orientação Sexual, Orientação Sexual Minoria

```python
# Verificando Valores
dados_funcionarios.groupby('Sexual_Orientation').size()
```

```python
# Traduzindo
dicionario_orientacao_sexual = {
    'Heterosexual': 'Heterossexual',
    'Bisexual': 'Bissexual',
    'Homosexual': 'Homossexual',
    'Prefer not to say': 'Prefere Não Informar',
    'Other': 'Outro'
}
dados_funcionarios['Orientação Sexual'] = dados_funcionarios['Sexual_Orientation'].map(dicionario_orientacao_sexual)
```

```python
# Visto que no estudo não haviam as opções pansexual e assexual - vamos substituir alguns outros por assexual, e alguns bissexuais por pansexual

# Substituição de Outros por Assexual (20%)
outros = dados_funcionarios['Sexual_Orientation'] == 'Other'
n_outros = outros.sum()
valores_outros = ['Assexual'] * (n_outros//5) + ['Outro'] * (n_outros-(n_outros//5))
np.random.shuffle(valores_outros)
dados_funcionarios.loc[outros, 'Orientação Sexual'] = valores_outros
dados_funcionarios.loc[outros].groupby('Orientação Sexual').size()
```

```python
# Substituição de BissexuaL por Pansexual (40%)
bissexuais = dados_funcionarios['Sexual_Orientation'] == 'Bisexual'
n_bissexuais = bissexuais.sum()
valores_bissexuais = ['Pansexual'] * ((n_bissexuais//5)*2) + ['Bissexual'] * (n_bissexuais-((n_bissexuais//5)*2))
np.random.shuffle(valores_bissexuais)
dados_funcionarios.loc[bissexuais, 'Orientação Sexual'] = valores_bissexuais
dados_funcionarios.loc[bissexuais].groupby('Orientação Sexual').size()
```

```python
dados_funcionarios.groupby('Orientação Sexual').size()
```

```python
# Definindo a Minoria do Grupo
dados_funcionarios['Orientação Sexual Minoria'] = 'Heterossexual'
dados_funcionarios.loc[dados_funcionarios['Orientação Sexual']!='Heterossexual', 'Orientação Sexual Minoria'] = 'Não Heterossexual' # outro é considerado não hétero
dados_funcionarios.loc[dados_funcionarios['Orientação Sexual']=='Prefere Não Informar', 'Orientação Sexual Minoria'] = 'Prefere Não Informar'
```

#### Salário

```python
faixas_subordinado = {
    'TI': (3500, 5500),
    'Marketing': (2000, 3200),
    'Financeiro': (2200, 3500),
    'Recursos Humanos': (2000, 3200),
    'Pesquisa e Desenvolvimento': (3200, 5000),
    'Vendas': (1800, 3000),
    'Operações': (1600, 2800),
}

faixas_gerente = {
    'TI': (13000, 18000),
    'Marketing': (10000, 15000),
    'Financeiro': (11000, 16000),
    'Recursos Humanos': (9000, 13500),
    'Pesquisa e Desenvolvimento': (12000, 17000),
    'Vendas': (10000, 15000),
    'Operações': (10000, 14000),
}

def gerar_salario(row):
    cargo = row["Cargo"]
    dep = row["Departamento"]

    if cargo=="Subordinado":
        lim_inf, lim_sup = faixas_subordinado[dep]
    else:
        lim_inf, lim_sup = faixas_gerente[dep]

    return np.random.uniform(lim_inf, lim_sup)

dados_funcionarios["Salário"] = (dados_funcionarios.apply(gerar_salario, axis=1).round(2))
```

```python
# Diminuindo salários de algumas minorias
# 0 = sem desconto
dados_funcionarios['Fator_Desconto'] = 1.0

dados_funcionarios.loc[dados_funcionarios['Gênero Social'] == 'Mulher', 'Fator_Desconto'] *= 0.97
dados_funcionarios.loc[dados_funcionarios['PCD'] == 'Sim', 'Fator_Desconto'] *= 0.96
dados_funcionarios.loc[dados_funcionarios['Etnia Minoria'] == 'Preta/Parda/Indígena', 'Fator_Desconto'] *= 0.94

dados_funcionarios['Salário'] = dados_funcionarios['Salário'] * dados_funcionarios['Fator_Desconto']
```

#### Grau de escolaridade

```python
quantis = dados_funcionarios['Salário'].quantile([0.19, 0.69, 0.89, 1.00])

q_fundamental_max = quantis[0.19]
q_medio_max = quantis[0.69]
q_superior_max = quantis[0.89]
def atribuir_escolaridade_quantil(row):
    salario = row['Salário']

    # Prioridade 1: Salários muito altos (Top 10%)
    if salario > q_superior_max:
        return 'Pós-Graduação/Especialização'

    # Prioridade 2: Salários altos (70% a 89%)
    elif salario > q_medio_max:
        return 'Superior Completo'

    # Prioridade 3: Salários médios (20% a 69%)
    elif salario > q_fundamental_max:

        # AJUSTE FINO (Opcional, mas recomendado para realismo)
        # Se for um Gerente, damos um pequeno boost de chance para Superior
        if row['Cargo'] == 'Gerente' or row['Etnia'] == 'Branca':
            return np.random.choice(
                ['Médio Completo', 'Superior Completo'],
                p=[0.75, 0.25] # 25% de chance de subir para Superior
            )
        return 'Médio Completo'

    # Prioridade 4: Salários baixos (0% a 19%)
    else:

        # AJUSTE FINO (Opcional)
        # Se for uma minoria, damos um pequeno boost de chance para não ficar no nível mais baixo
        if row['Grau Minoria'] > 2 and row['Etnia'] not in ['Branca']:
            return np.random.choice(
                ['Fundamental Completo', 'Médio Completo'],
                p=[0.85, 0.15] # 15% de chance de subir para Médio
            )
        return 'Fundamental Completo'


dicio_niveis_escolaridade = {
    'Fundamental Completo':1,
    'Médio Completo':2,
    'Superior Completo':3,
    'Pós-Graduação/Especialização':4
}
dados_funcionarios['Grau de escolaridade'] = dados_funcionarios.apply(atribuir_escolaridade_quantil, axis=1)
dados_funcionarios['Número Grau de escolaridade'] = dados_funcionarios['Grau de escolaridade'].map(dicio_niveis_escolaridade)
```
